```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
file = "RAW_data/Public_Assistance_Summary.xlsx"
```

```python
# Read our Data file for 2019 with the pandas library
data = pd.read_excel(file)
data_df = pd.DataFrame(data)
data_df
```

```python
# print out column names
for column in data_df.columns:
    print(column)
```

```python
modifiedData = data_df[['disasterNumber', 
                        'declarationDate', 
                        'Year', 
                        'incidentType',
                        'dcc',
                        'damageCategory',
                        'county',
                        'state',
                        'totalObligated',
                        'obligatedDate']]
modifiedData
```

```python
#Dropping years
drop2010 = modifiedData[modifiedData.Year != 2010]
drop2011 = drop2010[drop2010.Year != 2011]
drop2012 = drop2011[drop2011.Year != 2012]
drop2013 = drop2012[drop2012.Year != 2013]
drop2014 = drop2013[drop2013.Year != 2014]
drop2015 = drop2014[drop2014.Year != 2015]
finalYears = drop2015[drop2015.Year != 2020]
finalYears
```

```python
#confirm years are what we wante (2016-2019)
finalYears.Year.unique()
```

```python
#find all incident types
finalYears.incidentType.unique()
```

```python
#Dropping incident types
dropFire = finalYears[finalYears.incidentType != 'Fire']
dropFlood = dropFire[dropFire.incidentType != 'Flood']
dropIce = dropFlood[dropFlood.incidentType != 'Severe Ice Storm']
dropStorm = dropIce[dropIce.incidentType != 'Severe Storm(s)']
dropSnow = dropStorm[dropStorm.incidentType != 'Snow']
dropCoast = dropSnow[dropSnow.incidentType != 'Coastal Storm']
dropMud = dropCoast[dropCoast.incidentType != 'Mud/Landslide']
finalIncident = dropMud[dropMud.incidentType != "Typhoon"]
finalIncident
```

```python
#Count disasters incidents for each state (DESIRED YEARS AND INCIDENTS)
disasterData = finalIncident[['disasterNumber',  
                        'Year', 
                        'incidentType',
                        'state']]
countDisaster = disasterData.groupby(['Year','state','incidentType']).count()
countDisaster
```

```python
#Total Obligates amounts by FEMA by State (DESIRED YEARS AND INCIDENTS)
obligated = finalIncident[['totalObligated',  
                        'Year', 
                        'incidentType',
                        'state']]
totalObligated = obligated.groupby(['Year','state','incidentType']).sum()
totalObligated
```

```python
#Count disasters incidents for each state (DESIRED YEARS AND ALL INCIDENTS)
disasterData = finalYears[['disasterNumber',  
                        'Year', 
                        'incidentType',
                        'state']]
countDisaster = disasterData.groupby(['Year','state','incidentType']).count()
countDisaster
```

```python
#Total Obligates amounts by FEMA by State (DESIRED YEARS AND ALL INCIDENTS)
obligated = finalYears[['totalObligated',  
                        'Year', 
                        'incidentType',
                        'state']]
totalObligated = obligated.groupby(['Year','state','incidentType']).sum()
totalObligated
```
